```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

one_round_differentials = pd.read_csv("1_1_round_differentials").drop("Unnamed: 0", axis=1)
two_round_differentials = pd.read_csv("2_round_differentials").drop("Unnamed: 0", axis=1)
three_round_differentials = pd.read_csv("3_round_differentials").drop("Unnamed: 0", axis=1)
```

```python
one_round_differentials
```

```python
fig, axis = plt.subplots(2, 2, figsize=(10, 8))
axis = axis.flatten()
for i in range(4):
    sns.histplot(data=one_round_differentials, x=f"Round{i}", discrete=True, ax=axis[i])
    
plt.show()
```

```python
fig, axis = plt.subplots(2, 2, figsize=(10, 8))
axis = axis.flatten()
for i in range(4):
    sns.histplot(data=two_round_differentials, x=f"Round{i}", discrete=True, ax=axis[i])
    
plt.show()
```

```python
fig, axis = plt.subplots(2, 2, figsize=(10, 8))
axis = axis.flatten()
for i in range(4):
    sns.histplot(data=three_round_differentials, x=f"Round{i}", discrete=True, ax=axis[i])
    
plt.show()
```
