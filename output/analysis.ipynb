```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

one_round_differentials = pd.read_csv("1_round_differentials").drop("Unnamed: 0", axis=1)
two_round_differentials = pd.read_csv("2_round_differentials").drop("Unnamed: 0", axis=1)
three_round_differentials = pd.read_csv("3_round_differentials").drop("Unnamed: 0", axis=1)
four_round_differentials = pd.DataFrame(np.minimum(pd.read_csv("4_round_differentials").drop("Unnamed: 0", axis=1).values,
                                 pd.read_csv("4_round_differentials_low").drop("Unnamed: 0", axis=1).values),
                      index=four_round_differentials.index,
                      columns=four_round_differentials.columns).replace(-55, pd.NA)
```

```python
four_round_differentials.min()
```

```python
fig, axis = plt.subplots(2, 2, figsize=(10, 8))
axis = axis.flatten()
for i, ax in enumerate(axis):
    sns.histplot(data=one_round_differentials, x=f"Round{i}", discrete=True, ax=axis[i])
    ax.set_xticks(range(2, 7))
    ax.set_xlabel(f"$-log_2(p)$")
    
plt.show()
```

```python
fig, axis = plt.subplots(2, 2, figsize=(10, 8))
axis = axis.flatten()
for i, ax in enumerate(axis):
    sns.histplot(data=two_round_differentials, x=f"Round{i}", discrete=True, ax=axis[i])
    # ax.set_xlim(43, 56)
    ax.set_xticks(range(8, 19))
    ax.set_xlabel(f"$-log_2(p)$")
    
plt.show()
```
